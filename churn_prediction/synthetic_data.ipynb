```python
import pandas as pd
from sklearn.utils import shuffle
from imblearn.over_sampling import RandomOverSampler

# Split the DataFrame into two parts: one for columns to remain unchanged, and the other for columns to be shuffled
unchanged_df = unique_df.iloc[:,range(0,9)]
shuffled_df1 = unique_df.iloc[:,range(9,10)]
shuffled_df2 = unique_df.iloc[:5000,range(10,11)]
shuffled_df3 = unique_df.iloc[:,range(11,12)]



# Duplicate the rows to create multiple classes
n = 20
df_duplicates1 = pd.concat([shuffled_df1] * n, ignore_index=True).head(180000)
df_duplicates2 = pd.concat([shuffled_df2] * 40, ignore_index=True).head(180000)
df_duplicates3 = pd.concat([shuffled_df3] * n, ignore_index=True).head(180000)
df_duplicates = pd.concat([unchanged_df] * n, ignore_index=True).head(180000)


# Oversample the shuffled part of the DataFrame
ros = RandomOverSampler(random_state=None)
print('here')
X_sampled1, Y_sampled1 = ros.fit_resample(df_duplicates, df_duplicates1)
X_sampled2, Y_sampled2 = ros.fit_resample(df_duplicates, df_duplicates2)
# print('here')
X_sampled3, Y_sampled3 = ros.fit_resample(df_duplicates, df_duplicates3)



```

Resuffling of Data in columns

```python
# resuffling of data

shuffled_column2 = Y_sampled2['Product Code'].sample(frac=1).reset_index(drop=True)
shuffled_column1 = Y_sampled1['City'].sample(frac=1).reset_index(drop=True)
shuffled_column3 = Y_sampled3['Time Spent'].sample(frac=1).reset_index(drop=True)


# Assign the shuffled column back to the DataFrame
Y_sampled2['Product Code'] = shuffled_column2
Y_sampled1['City'] = shuffled_column1
Y_sampled3['Time Spent'] = shuffled_column3
```

```python
# adding back to the final_df

final_df = pd.concat([X_sampled1.head(180000),Y_sampled1.head(180000),Y_sampled2.head(180000),Y_sampled3.head(180000)],axis=1)
final_df.reset_index(drop=True, inplace=True)
final_df
```

```python
# dataframe to csv conversion with index not as a column
final_df.to_csv('synthetic_data.csv', index=False)
```
