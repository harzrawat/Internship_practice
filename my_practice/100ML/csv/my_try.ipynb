# Sep parameter

While reading a tsv(tab separated values) file, we use pd.read_csv but bcz by default sep = , for csv so we need to change it to '/'

Reading tsv format

```python
import pandas as pd

df = pd.read_csv('mytsv.tsv', sep='\t')  # Note: We're using sep='\t' because it's a TSV file
print(df.head())
```

# use_cols parameter

```python
pd.read_csv('dirty_data.csv',usecols=['City','Transaction_Amount']).head()
```

# Skip rows/nrows parameter

```python
pd.read_csv('dirty_data.csv',skiprows=[1,2]).head()
# skiprows=[1,2]  here 1,2 are row number starting from 0
```

In this parameter you can use function also

```python
df = pd.read_csv('dirty_data.csv').head()
print(df)
func = lambda x: x**2
pd.read_csv('dirty_data.csv',skiprows= func(2)).head()
# can pass any function as value of skip rows
```

```python
pd.read_csv('dirty_data.csv',nrows=2).head()
```

# Encoding Parameter

```python
df = pd.read_csv('dirty_data.csv').head()
df.to_csv('encoded.csv',encoding='iso-8859-1',index=False)
```

```python
pd.read_csv('encoded.csv')
```

```python
import pandas as pd

# Read the entire CSV file
df = pd.read_csv('dirty_data.csv')

# Write the entire DataFrame to a new CSV with different encoding
df.to_csv('encoded.csv', encoding='iso-8859-1', index=False)

# Optionally, you can print the first few rows of both files to compare
print("Original file:")
print(pd.read_csv('dirty_data.csv').head())
print("\nEncoded file:")
pd.read_csv('encoded.csv', encoding='iso-8859-1').head()
```
