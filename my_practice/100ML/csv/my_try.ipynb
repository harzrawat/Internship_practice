# Sep parameter

While reading a tsv(tab separated values) file, we use pd.read_csv but bcz by default sep = , for csv so we need to change it to '/'

```python
import pandas as pd

url = 'https://gist.githubusercontent.com/cdroulers/1a919d7f9ce701a716b0/raw/sample.tsv'
df = pd.read_csv(url, sep='\t')
print(df.head())
```

```python
import pandas as pd

url = 'https://gist.githubusercontent.com/cdroulers/1a919d7f9ce701a716b0/raw/sample.tsv'
df = pd.read_csv(url, sep='\t')  # Note: We're using sep='\t' because it's a TSV file
print(df.head())
```

```python
import pandas as pd
import requests
from io import StringIO

url = 'https://gist.githubusercontent.com/cdroulers/1a919d7f9ce701a716b0/raw/sample.tsv'
response = requests.get(url)
df = pd.read_csv(StringIO(response.text), sep='\t')
print(df.head())
```
