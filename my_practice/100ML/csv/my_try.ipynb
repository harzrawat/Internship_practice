# Skip rows/nrows parameter

```python
pd.read_csv('dirty_data.csv',skiprows=[1,2]).head()
# skiprows=[1,2]  here 1,2 are row number starting from 0
```

In this parameter you can use function also

```python
df = pd.read_csv('dirty_data.csv').head()
print(df)
func = lambda x: x*2
df_modified = pd.read_csv('dirty_data.csv',skiprows= [func(i) for i in range(5)]).head()
df_modified
# can pass any function as value of skip rows
```

```python
import pandas as pd

# Read the entire CSV file
df = pd.read_csv('dirty_data.csv')

# Write the entire DataFrame to a new CSV with different encoding
df.to_csv('encoded.csv', encoding='iso-8859-1', index=False)

# Optionally, you can print the first few rows of both files to compare
print("Original file:")
print(pd.read_csv('dirty_data.csv').head())
print("\nEncoded file:")
pd.read_csv('encoded.csv', encoding='iso-8859-1').head()
```

# Skipping Bad lines

Default behavior (on_bad_lines='error'):

can't be parsed will raise an exception, stopping the reading process.
Alternative options:

on_bad_lines='warn': Parses the rest of the file but prints a warning message for each line that couldn't be read.

on_bad_lines='skip': Simply skips the problematic lines and continues reading the rest of the file.

```python
import pandas as pd
pd.read_csv('bad_data.csv',sep=',',encoding='utf-8',on_bad_lines='skip')
```
